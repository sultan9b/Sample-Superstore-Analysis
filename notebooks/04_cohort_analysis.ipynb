## **Cohort Analysis (Когортный анализ)**
*Когортный анализ* - один из методов исследования потребителей.  <br>
**Цель:** определить как ведут себя клиенты в группах по одинаковым характеристикам или атрибутам и отслеживание их поведения во времени.

```python
# Импорт нужных библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import clickhouse_connect
import streamlit as st
```

```python
## Подключение к базе данных ClickHouse и загрузка данных в DataFrame
# Данные для подключения к ClickHouse
ch_host = 'localhost'
ch_port = '8123'
ch_user = os.getenv('CLICKHOUSE_USER')
ch_password = os.getenv('CLICKHOUSE_PASSWORD')
ch_database = 'dwh'

# Создание соединение с ClickHouse
ch_connect = clickhouse_connect.get_client(
    host=ch_host,
    port=ch_port,
    username=ch_user,
    password=ch_password,
    database=ch_database
)
```

```python
# Загрузка и чтение данных с clickhouse
df = ch_connect.query_df('SELECT * FROM dwh.sales_fact')
print("\033[1m" + 'ТАБЛИЦА ПО ПРОДАЖАМ ЗАГРУЖЕНА УСПЕШНО!' + "\033[1m" + '\n')
print(f'- Количество строк в таблице: {len(df)}', f'- Количество столбцов в таблице: {len(df.columns)}', sep='\n', end='\n\n')
df.head(10)
```

```python
# Вывод информации о клиентах и их заказах
print(f'| {"ТАБЛИЦА ПО ПРОДАЖАМ":<27}', '|')
print( '-' * 31)
print(f'| {"Количество заказов":<20} | {len(df.order_id.unique()):<3} |')
print(f'| {"Количество клиентов":<20} | {len(df.customer_name.unique()):<4} |')
print(f'| {"Количество сегментов":<15} | {len(df.segment.unique()):<4} |')
```
